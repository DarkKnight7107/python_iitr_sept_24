## Problem 6: Armstrong Number

*You are given an integer num. Return 'true' if num is an Armstrong number, and 'false' otherwise.*

An Armstrong number is a number (with 'k' digits) such that the sum of its digits raised to 'kth' power is equal to the number itself. For example, 371 is an Armstrong number because 3^3 + 7^3 + 1^3 = 371.

**Example 1:**
**Input:** num = 1

**Output:** True

**Explanation:**
1 is an Armstrong number as, 1^1 = 1.

**Example 2:**
**Input:** num = 1634

**Output:** True

**Explanation:**
1634 is an Armstrong number, as 1^4 + 6^4 + 3^4 + 4^4 = 1634.

## Problem 8: Sum of All Divisor

*You are given an integer num. Find the sum of sum_all_divisor(i) for all ‘i’ from 1 to num.*

**Example 1:**
**Input:** num  = 5

**Output:** 21

**Explanation:**
- We need to find the sum of ‘sum_all_divisors(i)’ for all ‘i’ from 1 to 5. 
- sum_all_divisor(1) = 1
- sum_all_divisor(2) = 2 + 1 = 3
- sum_all_divisor(3) = 3 + 1 = 4
- sum_all_divisor(4) = 4 + 2 +1 = 7 
- sum_all_divisor(5) = 5 + 1 = 6
- Therefore our answer is sum_all_divisor(1) + sum_all_divisor(2) + sum_all_divisor(3) + sum_all_divisor(4) + sum_all_divisor(5) = 1 + 3 + 4 + 7 + 6 = 21.
